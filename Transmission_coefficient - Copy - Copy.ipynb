```python
euler_number  = 2.71828

# m
#unit: kg
mass_electron = 9.10938 * (10 ** (-31))
mass_proton   = 1.67262 * (10 ** (-27))
mass_h2       = 1.67353 * (10 ** (-27)) * 2


# L
# unit: m
len_box       = 5.0 * (10 ** (-10))

# E 1ev
# unit: J
E_electron    = 1.60 * (10 ** (-19))
E_proton      = 1.60 * (10 ** (-19))
E_h2          = 3.20 * (10 ** (-19))

# h_slash
# unit: Js
planck_const  = 1.05457 * (10 ** (-34))

# V
# Unit: J
potential_energy = 'unknown'


def T_simple(epsilon, kL, euler_number=2.718281828):
    #print epsilon
    T = 16 * epsilon * (1-epsilon) * ( euler_number ** ((-2)*kL) )
    return T
#-----------------------------------------------------

def kL(mass, V, E, L, planck_const):
    return (2 * mass * abs(V-E)) ** (0.5) / planck_const * L
#-----------------------------------------------

def V(L, mass, planck_const, const):
    return (const ** 2) * (planck_const ** 2) / 2 / mass/ (L ** 2) 
#-----------------------------------------------

def T_real(epsilon, kL, euler_number=2.718281828):
    #print epsilon
    if epsilon == 0:
        return 0
    elif epsilon == 1:
        return 0
    else:
        return (1 + (euler_number ** kL - euler_number ** (-1 * kL))**2 / (16 * epsilon * (1-epsilon))) ** (-1)
#--------------------------------------------------------------

def T_E(epsilon, mass, E):
    if epsilon == 0:
        return 0
    V = E / epsilon
    #print V, E, mass, planck_const
    kL_v = kL(mass, V, E, len_box, planck_const)
    #print kL_v
    if kL_v > 10:
        return T_simple(epsilon, kL_v)
    else:
        return T_real(epsilon, kL_v)
#------------------------------------------

def T_V(epsilon, mass, V):
    if epsilon == 0:
        return 0
    E = V * epsilon
    #print V, E, mass, planck_const
    kL_v = kL(mass, V, E, len_box, planck_const)
    #print kL_v
    if kL_v > 10:
        return T_simple(epsilon, kL_v)
    else:
        return T_real(epsilon, kL_v)
#------------------------------------------

def T_E_list(mass, E, start, end, step, divisor=100):
    epsilon_list = []
    T_list = []
    for i in range(start, end, step):
        epsilon = i / float(divisor)
        epsilon_list.append(epsilon)
        T_list.append(T_E(epsilon, mass, E))   
    return epsilon_list, T_list
#----------------------------------------------    
def T_V_list(mass, V, start, end, step, divisor=100):
    epsilon_list = []
    T_list = []
    for i in range(start, end, step):
        epsilon = i / float(divisor)
        epsilon_list.append(epsilon)
        T_list.append(T_V(epsilon, mass, V))   
    return epsilon_list, T_list
#----------------------------------------------  
```

# Tansmission coefficient of electron

```python
%pylab inline
import matplotlib.pyplot as plt

# For electron
# When E < V

EV_epsilon_list, EV_T_list = T_list(mass_electron, E_electron, 0, 90, 1, 100)
#print EV_epsilon_list, EV_T_list

tmp_epsilon, tmp_T = T_list(mass_electron, E_electron, 900, 1000, 5, 1000)
#print tmp_epsilon, tmp_T

EV_epsilon_list.extend(tmp_epsilon)
EV_T_list.extend(tmp_T)  

#print EV_epsilon_list, EV_T_list

# When E > V

VE_epsilon_list, VE_T_list = T_list(mass_electron, E_electron, 100, 400, 1, 100)


#---------Plot-------
fig = plt.figure(figsize=(16,8))

plt.subplot(121)
plt.plot(EV_epsilon_list, EV_T_list)

plt.subplot(122)
plt.plot(VE_epsilon_list, VE_T_list)

plt.show()
```
