# Transmission coefficient of electron, proton, H2 giving different potential energy

```python
%pylab inline
# Natureal Numbers
euler_number  = 2.71828

# Mass of the Particles
# m
#unit: kg
mass_electron = 9.10938 * (10 ** (-31))
mass_proton   = 1.67262 * (10 ** (-27))
mass_h2       = 1.67353 * (10 ** (-27)) * 2


# Length of the barrier
# L
# unit: m
len_box       = 5.0 * (10 ** (-10))

# 
# E 1ev
# unit: J
E_electron    = 1.60 * (10 ** (-19))
E_proton      = 1.60 * (10 ** (-19))
E_h2          = 3.20 * (10 ** (-19))

# h_bar
# unit: Js
h_bar  = 1.05457 * (10 ** (-34))



#----------Definition of the Terms------------

def kL(mass, V, E, L, h_bar):
    '''
    Aim: To calculate the value of kL corresponding to (2m |V-E|) ^ (1/2) / h_bar* L
    Terms：
        mass：mass of the particles
        V：   potential energy
        E:    transmition energy
        L:    length of the barrier
        h_bar
    '''
    return (2 * mass * abs(V-E)) ** (0.5) / h_bar * L
#-----------------------------------------------


def T_real(epsilon, kL, euler_number=2.718281828):
    '''
    Aim： By using equation 8.19a to calculate Trnasmission probability for all kL (not only for kL》1).
         
    Terms： 
          epsilon: epsilon = E/V, the range of epsilon: 0-1
          kL:      the values gotten from function kL
          euler_number: e,2.71828
    '''
    
    if epsilon == 0:
        return 0
    elif epsilon == 1:
        return 0
    else:
        return (1 + (euler_number ** kL - euler_number ** (-1 * kL)) ** 2 / (16 * epsilon * (1-epsilon))) ** (-1)
#--------------------------------------------------------------
def T_simple(epsilon, kL, euler_number=2.718281828):
    '''
    Aim： By using equation 8.19b to calculate Trnasmission probability only for kL》1.
          
    Terms： 
          epsilon, kL, euler_number
    '''
    
    T = 16 * epsilon * (1-epsilon) * ( euler_number ** ((-2)*kL) )
    return T
#-----------------------------------------------------


def T_V(epsilon, mass, V):
    '''
    The integration of the three functions defined above. When V has fixed values, use this integration.
    Terms：
           epsilon, mass
           V:    the input potential energy
    '''
    if epsilon == 0:
        return 0
    
    #Calculate E from V and epsilon
    E = V * epsilon
    
    #Save the calculated kL to kL_v
    kL_v = kL(mass, V, E, len_box, h_bar)
    
    
    if kL_v > 10:
        return T_simple(epsilon, kL_v)
    
    else:
        return T_real(epsilon, kL_v)
#------------------------------------------


   
def T_V_list(mass, V, start, end, step, divisor=100):
    '''
    Aim：Set up for the function T_V, produce a list of epsilon values, get the coresponding T values.
    '''
    epsilon_list = []
    T_list = []
    for i in range(start, end, step):
        epsilon = i / float(divisor)
        epsilon_list.append(epsilon)
        T_list.append(T_V(epsilon, mass, V))   
    return epsilon_list, T_list
#----------------------------------------------  



def simulate_V(L, VL, massL, labelL):
    '''
    The main function for dealing with multiple particles.
        VL: list of V
        massL： list of particle masses (electron, proton, Hydrogen molecule)
        labelL: for labeling the related names to the masses
    '''
    #define a new group for saving the calculated values of T and epsilon  
    Vincent = {}
    
    Vincent['EV'] = {}
    #As E<V
    #only write aDict['EV'] = {}
    len_mass = len(massL)
    #Take every value from the list of particle masses
    for i in range(len_mass):
        mass = massL[i]
        label = labelL[i]
        #Calculate V
        V2 = VL[i]
        #Calulte epsilon and T through the intergration of the functions
        EV2_epsilon_list, EV2_T_list = T_V_list(mass, V2, 0, 90, 1, 100)
        tmp_epsilon, tmp_T = T_V_list(mass, V2, 900, 1000, 5, 1000)
        EV2_epsilon_list.extend(tmp_epsilon)
        EV2_T_list.extend(tmp_T)
        Vincent['EV'][label] = [EV2_epsilon_list, EV2_T_list]
        #---------------------------------------------
        #VE_epsilon_list, VE_T_list = T_V_list(mass, V2, 100, 400, 1, 100)
        #Vincent['VE'][label] = [VE_epsilon_list, VE_T_list]
    #-------------------------------------
    return Vincent
#-------------------------------------------------


def plot_lines(Vincent, title="title"):
    '''
    Aim： Plot
    Terms：
           Vincent: The collection the calculated values of epsilon and T by simulate_V.
           title：title of the plot
    '''
    fig = plt.figure(figsize=(16,7))

    plt.subplot(121)
    plt.xlabel(r'Epsilon, $\epsilon, E < V $')
    plt.ylabel(r'Transmission coefficient, $\mathit{T}$')
    plt.title(title)
    for const, valueL in Vincent['EV'].items():
        plt.plot(valueL[0], valueL[1], label=const)
    legend()

    plt.show()
#--------------------------------------------------


import matplotlib.pyplot as plt

V = raw_input("Please input the potential energy V (unit: eV) for electron, protron and H2 in the format like '0.1,0.0005,0.0008' and press enter\n")

VL = [float(i.strip()) * E_electron for i in V.split(',')]

Vincent = simulate_V(len_box, VL, [mass_electron, mass_proton, mass_h2], ['electron', 'proteon', 'H2']) 

plot_lines(Vincent, r"Transmission coefficient of electron, proton, H2")
```

```python
%pylab inline
import matplotlib.pyplot as plt

def V(L, mass, h_bar, const):
    '''
    目的：计算V的值，根据L * (2mV)^(1/2) / h_bar = 2 or 4 or 10.
    参数：
        L:    barrier的长度
        mass：粒子的质量值
        planck_const: 普朗克常数，h_bar
        const: 代表常量，即公式 L * (2mV)^(1/2) / h_bar 的值，根据图表取值为2,4,10
    '''
    return (const ** 2) * (h_bar ** 2) / 2 / mass/ (L ** 2) 
#-----------------------------------------------


def simulate_2(mass, L, constL):
    '''
    constL = [ L * (2mV) ** (0.5) / h, ... ]
    '''
    aDict = {}
    aDict['EV'] = {}
    aDict['VE'] = {}
    for const in constL:
        V2 = V(L, mass, h_bar, const)
        EV2_epsilon_list, EV2_T_list = T_V_list(mass, V2, 0, 90, 1, 100)
        tmp_epsilon, tmp_T = T_V_list(mass, V2, 900, 1000, 5, 1000)
        EV2_epsilon_list.extend(tmp_epsilon)
        EV2_T_list.extend(tmp_T)
        aDict['EV'][const] = [EV2_epsilon_list, EV2_T_list]
        #---------------------------------------------
        VE_epsilon_list, VE_T_list = T_V_list(mass, V2, 100, 400, 1, 100)
        aDict['VE'][const] = [VE_epsilon_list, VE_T_list]
    #-------------------------------------
    return aDict
#-------------------------------------------------

# For electron

aDict = simulate_2(mass_electron, len_box, [2,4,10])
plot_lines(aDict, "Transmission coefficient of electron")
```
