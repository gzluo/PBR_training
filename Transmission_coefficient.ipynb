```python
%pylab inline
# 自然常数
euler_number  = 2.71828

# 各种粒子的质量值
# m
#unit: kg
mass_electron = 9.10938 * (10 ** (-31))
mass_proton   = 1.67262 * (10 ** (-27))
mass_h2       = 1.67353 * (10 ** (-27)) * 2


# 盒子的长度
# L
# unit: m
len_box       = 5.0 * (10 ** (-10))

# 各种粒子的电量，H2应该是不带电的吧，不过电量的值最后也没用着。
# E 1ev
# unit: J
E_electron    = 1.60 * (10 ** (-19))
E_proton      = 1.60 * (10 ** (-19))
E_h2          = 3.20 * (10 ** (-19))

# 普朗克常量，在公式中是h加一短线
# h_bar
# unit: Js
planck_const  = 1.05457 * (10 ** (-34))

# 势能，未知，后面有公式计算势能
# V
# Unit: J
potential_energy = 'unknown'

#----------下面是定义的操作函数，是对方程的操作------------


def kL(mass, V, E, L, planck_const):
    '''
    目的：计算kL的值，根据公式 (2m |V-E|) ^ (1/2) / h_bar* L
    参数：
        mass：粒子的质量值
        V：   势能
        E:    透射能
        L:    barrier的长度
        planck_const: 普朗克常数，h_bar
    '''
    return (2 * mass * abs(V-E)) ** (0.5) / planck_const * L
#-----------------------------------------------


def T_real(epsilon, kL, euler_number=2.718281828):
    '''
    目的： 套用公式8.19a计算 Trnasmission probability。不论kL是否远大于1都可用。
          但通常kL的值不能大于64，否则得出的数会特别大 （2.71828的63次方，2.71828^64）。
    参数： 
          epsilon: E/V的值，在区间0-1和1-4之间均匀间隔取出
          kL:      函数kL计算得到的值。
          euler_number: 自然对数的底数的值e，2.71828
    '''
    
    if epsilon == 0:
        return 0
    elif epsilon == 1:
        return 0
    else:
        return (1 + (euler_number ** kL - euler_number ** (-1 * kL)) ** 2 / (16 * epsilon * (1-epsilon))) ** (-1)
#--------------------------------------------------------------
def T_simple(epsilon, kL, euler_number=2.718281828):
    '''
    目的： 套用公式8.19b计算 Trnasmission probability。只在kL远大于1时可用。
          
    参数： 
          epsilon: E/V的值，在区间0-1和1-4之间均匀间隔取出
          kL:      函数kL计算得到的值。
          euler_number: 自然对数的底数的值e，2.71828
    '''
    
    T = 16 * epsilon * (1-epsilon) * ( euler_number ** ((-2)*kL) )
    return T
#-----------------------------------------------------


def T_V(epsilon, mass, V):
    '''
    目的： 这是一个整合函数，调用了前面定义的3个函数。在V固定的时候，使用这个函数。
    参数：
           epsilon: E/V的值，在区间0-1和1-4之间均匀间隔取出
           mass：粒子的质量值
           V:    barrier的势能
    '''
    if epsilon == 0:
        return 0
    
    #计算得到投射能 E
    E = V * epsilon
    
    #计算得到kL的值，存入kL_v
    kL_v = kL(mass, V, E, len_box, planck_const)
    #print kL_v
    #如果kL大于10，则调用简化的公式计算T的值
    if kL_v > 10:
        return T_simple(epsilon, kL_v)
    #如果kL小于10，则调用原始的公式计算T的值
    else:
        return T_real(epsilon, kL_v)
#------------------------------------------


   
def T_V_list(mass, V, start, end, step, divisor=100):
    '''
    目的：对函数T_V的封装，生成一些列的epsilon的值，并计算得到T值。
    '''
    epsilon_list = []
    T_list = []
    for i in range(start, end, step):
        epsilon = i / float(divisor)
        epsilon_list.append(epsilon)
        T_list.append(T_V(epsilon, mass, V))   
    return epsilon_list, T_list
#----------------------------------------------  



def simulate_V(L, VL, massL, labelL):
    '''
    目的：开始计算，通过这个函数调用之前定义过的函数，来计算T和epsilon的值。
    参数：
        VL: V的值, 一个列表，包含了电子、质子和氢分子的所处空间的势能
        massL： 一个列表，包含了电子、质子和氢分子的质量值
        labelL: 一个标签，顺序与massL一样，用以标记massL中质量值所属的粒子的名字
    '''
    aDict = {}
    #EV 表示 E<V的值
    aDict['EV'] = {}
    #VE 表示E>V的值
    #aDict['VE'] = {}
    len_mass = len(massL)
    #从列表中逐一取出质量值
    for i in range(len_mass):
        mass = massL[i]
        label = labelL[i]
        #计算势能V
        V2 = VL[i]
        #调用函数得到T值和epsilon值
        EV2_epsilon_list, EV2_T_list = T_V_list(mass, V2, 0, 90, 1, 100)
        tmp_epsilon, tmp_T = T_V_list(mass, V2, 900, 1000, 5, 1000)
        EV2_epsilon_list.extend(tmp_epsilon)
        EV2_T_list.extend(tmp_T)
        aDict['EV'][label] = [EV2_epsilon_list, EV2_T_list]
        #---------------------------------------------
        #VE_epsilon_list, VE_T_list = T_V_list(mass, V2, 100, 400, 1, 100)
        #aDict['VE'][label] = [VE_epsilon_list, VE_T_list]
    #-------------------------------------
    return aDict
#-------------------------------------------------


def plot_lines(aDict, title="title"):
    '''
    目的： 做出线图
    参数：
           aDict: 存储有函数simulate_V的返回值，即计算的epsilon和T的值
           title：图形的标题
    '''
    fig = plt.figure(figsize=(16,7))

    plt.subplot(121)
    plt.xlabel(r'Epsilon, $\epsilon, E < V $')
    plt.ylabel(r'Transmission coefficient, $\mathit{T}$')
    plt.title(title)
    for const, valueL in aDict['EV'].items():
        plt.plot(valueL[0], valueL[1], label=const)
    legend()

    plt.show()
#--------------------------------------------------


import matplotlib.pyplot as plt

V = raw_input("Please input the potential energy V (unit: eV) for electron, protron and H2 in the format like '0.1,0.0005,0.0008' and press enter\n")

VL = [float(i.strip()) * E_electron for i in V.split(',')]

aDict = simulate_V(len_box, VL, [mass_electron, mass_proton, mass_h2], ['electron', 'proteon', 'H2']) 

plot_lines(aDict, r"Transmission coefficient of electron, proton, H2")
```

```python
# 自然常数
euler_number  = 2.71828

# 各种粒子的质量值
# m
#unit: kg
mass_electron = 9.10938 * (10 ** (-31))
mass_proton   = 1.67262 * (10 ** (-27))
mass_h2       = 1.67353 * (10 ** (-27)) * 2


# 盒子的长度
# L
# unit: m
len_box       = 5.0 * (10 ** (-10))

# 各种粒子的电量，H2应该是不带电的吧，不过电量的值最后也没用着。
# E 1ev
# unit: J
E_electron    = 1.60 * (10 ** (-19))
E_proton      = 1.60 * (10 ** (-19))
E_h2          = 3.20 * (10 ** (-19))

# 普朗克常量，在公式中是h加一短线
# h_bar
# unit: Js
planck_const  = 1.05457 * (10 ** (-34))

# 势能，未知，后面有公式计算势能
# V
# Unit: J
potential_energy = 'unknown'

#----------下面是定义的操作函数，是对方程的操作------------

def V(L, mass, planck_const, const):
    '''
    目的：计算V的值，根据L * (2mV)^(1/2) / h_bar = 2 or 4 or 10.
    参数：
        L:    barrier的长度
        mass：粒子的质量值
        planck_const: 普朗克常数，h_bar
        const: 代表常量，即公式 L * (2mV)^(1/2) / h_bar 的值，根据图表取值为2,4,10
    '''
    return (const ** 2) * (planck_const ** 2) / 2 / mass/ (L ** 2) 
#-----------------------------------------------


def kL(mass, V, E, L, planck_const):
    '''
    目的：计算kL的值，根据公式 (2m |V-E|) ^ (1/2) / h_bar* L
    参数：
        mass：粒子的质量值
        V：   势能
        E:    透射能
        L:    barrier的长度
        planck_const: 普朗克常数，h_bar
    '''
    return (2 * mass * abs(V-E)) ** (0.5) / planck_const * L
#-----------------------------------------------


def T_real(epsilon, kL, euler_number=2.718281828):
    '''
    目的： 套用公式8.19a计算 Trnasmission probability。不论kL是否远大于1都可用。
          但通常kL的值不能大于64，否则得出的数会特别大 （2.71828的63次方，2.71828^64）。
    参数： 
          epsilon: E/V的值，在区间0-1和1-4之间均匀间隔取出
          kL:      函数kL计算得到的值。
          euler_number: 自然对数的底数的值e，2.71828
    '''
    
    if epsilon == 0:
        return 0
    elif epsilon == 1:
        return 0
    else:
        return (1 + (euler_number ** kL - euler_number ** (-1 * kL)) ** 2 / (16 * epsilon * (1-epsilon))) ** (-1)
#--------------------------------------------------------------
def T_simple(epsilon, kL, euler_number=2.718281828):
    '''
    目的： 套用公式8.19b计算 Trnasmission probability。只在kL远大于1时可用。
          
    参数： 
          epsilon: E/V的值，在区间0-1和1-4之间均匀间隔取出
          kL:      函数kL计算得到的值。
          euler_number: 自然对数的底数的值e，2.71828
    '''
    
    T = 16 * epsilon * (1-epsilon) * ( euler_number ** ((-2)*kL) )
    return T
#-----------------------------------------------------

def T_E(epsilon, mass, E):
    '''
    目的： 这是一个整合函数，调用了前面定义的3个函数。在E固定的时候，使用这个函数。
    参数：
           epsilon: E/V的值，在区间0-1和1-4之间均匀间隔取出
           mass：粒子的质量值
           E:    例子的投射能
    '''
    if epsilon == 0:
        return 0
    
    #计算得到势能V的值
    V = E / epsilon
    
    #计算得到kL的值，存入kL_v
    kL_v = kL(mass, V, E, len_box, planck_const)
    
    #如果kL大于10，则调用简化的公式计算T的值
    if kL_v > 10:
        return T_simple(epsilon, kL_v)
    #如果kL小于10，则调用原始的公式计算T的值
    else:
        return T_real(epsilon, kL_v)
#------------------------------------------

def T_V(epsilon, mass, V):
    '''
    目的： 这是一个整合函数，调用了前面定义的3个函数。在V固定的时候，使用这个函数。
    参数：
           epsilon: E/V的值，在区间0-1和1-4之间均匀间隔取出
           mass：粒子的质量值
           V:    barrier的势能
    '''
    if epsilon == 0:
        return 0
    
    #计算得到投射能 E
    E = V * epsilon
    
    #计算得到kL的值，存入kL_v
    kL_v = kL(mass, V, E, len_box, planck_const)
    
    #如果kL大于10，则调用简化的公式计算T的值
    if kL_v > 10:
        return T_simple(epsilon, kL_v)
    #如果kL小于10，则调用原始的公式计算T的值
    else:
        return T_real(epsilon, kL_v)
#------------------------------------------

def T_E_list(mass, E, start, end, step, divisor=100):
    '''
    目的：对函数T_E的封装，生成一些列的epsilon的值，并计算得到T值。
    '''
    epsilon_list = []
    T_list = []
    for i in range(start, end, step):
        epsilon = i / float(divisor)
        epsilon_list.append(epsilon)
        T_list.append(T_E(epsilon, mass, E))   
    return epsilon_list, T_list
#----------------------------------------------    
def T_V_list(mass, V, start, end, step, divisor=100):
    '''
    目的：对函数T_V的封装，生成一些列的epsilon的值，并计算得到T值。
    '''
    epsilon_list = []
    T_list = []
    for i in range(start, end, step):
        epsilon = i / float(divisor)
        epsilon_list.append(epsilon)
        T_list.append(T_V(epsilon, mass, V))   
    return epsilon_list, T_list
#----------------------------------------------  



def simulate_V(L, const, massL, labelL):
    '''
    目的：开始计算，通过这个函数调用之前定义过的函数，来计算T和epsilon的值。
    参数：
        L: barrier的长度
        const： 图中的2,4,10
        massL： 一个列表，包含了电子、质子和氢分子的质量值
        labelL: 一个标签，顺序与massL一样，用以标记massL中质量值所属的粒子的名字
    '''
    aDict = {}
    #EV 表示 E<V的值
    aDict['EV'] = {}
    #VE 表示E>V的值
    aDict['VE'] = {}
    len_mass = len(massL)
    #从列表中逐一取出质量值
    for i in range(len_mass):
        mass = massL[i]
        label = labelL[i]
        #计算势能V
        V2 = V(L, mass, planck_const, const)
        print V2
        #调用函数得到T值和epsilon值
        EV2_epsilon_list, EV2_T_list = T_V_list(mass, V2, 0, 90, 1, 100)
        tmp_epsilon, tmp_T = T_V_list(mass, V2, 900, 1000, 5, 1000)
        EV2_epsilon_list.extend(tmp_epsilon)
        EV2_T_list.extend(tmp_T)
        aDict['EV'][label] = [EV2_epsilon_list, EV2_T_list]
        #---------------------------------------------
        VE_epsilon_list, VE_T_list = T_V_list(mass, V2, 100, 400, 1, 100)
        aDict['VE'][label] = [VE_epsilon_list, VE_T_list]
    #-------------------------------------
    return aDict
#-------------------------------------------------


def plot_lines(aDict, title="title"):
    '''
    目的： 做出线图
    参数：
           aDict: 存储有函数simulate_V的返回值，即计算的epsilon和T的值
           title：图形的标题
    '''
    fig = plt.figure(figsize=(16,7))

    plt.subplot(121)
    plt.xlabel(r'Epsilon, $\epsilon, E < V $')
    plt.ylabel(r'Transmission coefficient, $\mathit{T}$')
    plt.title(title)
    for const, valueL in aDict['EV'].items():
        plt.plot(valueL[0], valueL[1], label=const)
    legend()

    plt.subplot(122)
    plt.xlabel(r'Epsilon, $\epsilon, E > V $')
    plt.ylabel(r'Transmission coefficient, $\mathit{T}$')
    plt.title(title)
    for const, valueL in aDict['VE'].items():
        plt.plot(valueL[0], valueL[1], label=const)
    legend()
    plt.show()
#--------------------------------------------------
```
